# GTEx combination

```python
GTEx_muscle = pd.read_table("GTEx data/gene_reads_muscle_skeletal.txt")
GTEx_muscle.GeneID = GTEx_muscle.GeneID.str.extract(r'(ENSG.+(?=\.))')
GTEx_muscle.drop(['GeneSymbol'], axis=1, inplace=True)
GTEx_muscle
```

# Helsinki combination

```python
Helsinki_muscle_L = pd.read_csv("Helsinki data/merge_L_Helsinki.csv")
# remove non-qualified samples (cell-line, mouse muscles) from the original merge_L_Helsinki.csv file
Meta_Helsinki = pd.read_excel('Meta/RNA-seq integration meta.xlsx', sheet_name='Helsinki')  
Helsinki_muscle = pd.concat([ Helsinki_muscle_L['GeneID'], Helsinki_muscle_L[Meta_Helsinki.Sample_id] ], axis=1)
Helsinki_muscle.GeneID = Helsinki_muscle.GeneID.str.extract(r'(ENSG.+(?=\.))')
Helsinki_muscle
```

# All combination

```python
All_muscle = pd.merge(GEO_muscle, GTEx_muscle, how='inner', on='GeneID')
All_muscle = pd.merge(All_muscle, Helsinki_muscle, how='inner', on='GeneID')
All_muscle
```

## Filter those genes > 0 in every samples 
- To remove those genes with count inflantion

```python
All_muscle = All_muscle.dropna(how='any').set_index('GeneID').astype(int) #remove one raw contains some NA
All_muscle = All_muscle[(All_muscle > 0).all(axis=1)] 
All_muscle = All_muscle.reset_index()
All_muscle
```

## Gene symbal dictionary
- Change the first column from Ensembl IDs to Gene symbols

```python
# load dictionary
GeneID_anno = pd.read_csv("Meta/Annotation_dict_from_GTEx.csv") # this file is extracted from the GTEx data
Genelist = GeneID_anno[GeneID_anno.GeneID.isin(All_muscle.GeneID)].GeneID.tolist()
GeneID_dict = dict(zip(GeneID_anno.GeneID, GeneID_anno.GeneSymbol))
```

```python
All_muscle['Symbol'] = All_muscle.GeneID.map(GeneID_dict)
All_muscle
```

```python
All_muscle.drop_duplicates(subset='Symbol', keep="first", inplace=True)
```
